Die wichtigsten Elemente von Klassen

````python
class KlassenName:
    # Die __init__ Funktion ist speziell und wird von Python automatisch ausgeführt, wenn wir 
    # mit dem Aufruf der Klasse eine neue Instanz erzeugen.
    def __init__(self, parameter1, parameter2, übergebener_wert_3):
        # Üblicherweise werden die bei __init__() übergebenen Werte gleich in der Instanz 
        # gespeichert und oft heissen die Funktionsparameter bei __init__() auch gleich oder 
        # ähnlich wie die Instanz-Variablen.
        self.parameter1 = parameter1
        self.parameter2 = parameter2
        # Die Parameternamen und die Namen der Variablen innerhalb der Klasse haben aber
        # nicht direkt etwas miteinander zu tun und können ganz unterschiedlich heissen
        self.parameter3 = übergebener_wert_3
    
    # Alle Funktionen (auch Methoden genannt) einer Klasse müssen immer als erstes einen 
    # `self`-Parameter entgegen nehmen. Dieser gibt uns Zugriff auf die Instanz.
    def methode(self, parameter_a):
        ...
        # Innerhalb dieser Funktion könnte auch auf die Variablen `self.parameter1`, 
        # `self.parameter2` und `self.parameter3` zugegriffen werden.
        

instanz_1 = KlassenName(12, 44, 12)  # Die Werte beim Aufruf der Klasse werden der Funktion 
# __init__ übergeben. Verschiedene Instanzen (auch Objekte genannt) sind grundsätzlich
# unabhängig voneinander und haben ihre eigenen Daten. Sie teilen aber die gemeinsamen 
# Funktionen (Methoden genannt) aus der Klasse.
instanz_1.methode("kajsghas")
instanz_2 = KlassenName(352, 23, 512)
instanz_2.methode("asfd")

# Die letzte Zeile ist im Grunde eine verkürzte Schreibweise für
KlassenName.methode(instanz_2, "asfd")
# Das wir es bei instanz_2 aber mit einer Instanz von `KlassenName` zu tun haben, ist
# bereits in instanz_2 enthalten. Wenn wir also KlassenName nochmal erwähnen, um die
# Methode `methode` aufzurufen, machen wir uns nur unnötig Arbeit.
````

Qualität und Korrektheit des Codes sicherzustellen gehört zu einer wichtigen Aufgabe bei der Softwareentwicklung. Ein Teil dieser Aufgabe findet in Form von Integrations und Systemtests auf Stufe von teil- oder vollintegrierten Systemen statt. Mit der Qualitätssicherung kann aber auch schon bei der täglichen Entwicklung angesetzt werden. Jede Funktion und jede Klasse die geschrieben wird kann schon während der Entwicklung mit Tests abgedeckt werden. Dies hat mehrere Vorteile

 * So entstandene Tests können immer wieder auch automatisiert ausgeführt werden und sicherstellen, dass keine Verschlechterung eingeführt wird.
 * Der Code wird schon so entwickelt, dass er auch gut testbar ist. Dies hat häufig auch positive Effekte auf das Design<sup>[Citation Needed]</sup> der Funktionen und Klassen.
 * Der Sichtwechsel für das Entwickeln von Tests hilft, auch schon während der Entwicklung über bestimmte Fehler- und Problemfälle nachzudenken.

```python
# Zuerst müssen wir unser Modul bzw. die gewünschte(n) Funktion(en) davon importieren. Obwohl die Datei die 
# Endung .py haben muss, lassen wir diesen Teil beim `import` weg.
#
# !!!! Dieser Code funktioniert innerhalb des Jupyter-Notebooks sehr wahrscheinlich nicht. Es kann funktionieren,

from mein_modul import meine_funktion

def test_meine_funktion_macht_abc():
   resultat = meine_funktion(1, 2, 3)
   assert resultat == 42


def test_meine_funktion_macht_xyz():
   resultat = meine_funktion("a", "b")
   assert resultat == "hello"
```

In die Testdatei schreiben wir nun unsere Test-Funktionen. Jede Test-Funktion muss auch wieder mit dem Namen `test_` beginnen. Es ist zu empfehlen, diese Funktion mit einem einigermassen sprechenden Namen zu versehen, der beschreibt, was der Test überhaupt prüft. Test-Funktionen können teilweise etwas längere Namen haben.

Die zu testende Funktion sollte irgendeinen Effekt haben wenn wir sie aufrufen. Dass hier der richtige und erwartete Effekte eingetreten ist, können wir mit dem `assert`-Statement prüfen.

## Test driven development (TDD)

Eine spezielle Art um Tests zu entwickeln ist die Methodik **TDD** - *test driven development*. Die Idee dahinter ist, die Code und die dazu gehörigen Tests Schritt für Schritt entstehen zu lassen. Die wichtigsten Prinzipien bei TDD sind

 * **Der Test wird zuerst geschrieben.** Man soll sich vor der Implementierung überlegen, wie das erwartete Verhalten aussehen soll.
 * **Vor der Implementierung muss der Test fehlschlagen.** Ein neuer Test soll gleich ausgeführt werden bevor überhaupt irgendwelcher Code geschrieben wird. Erst wenn ein Test fehlschlägt wissen wir, dass der Test auch wirklich zuschlagen würde, wenn wir später wieder etwas kaputt machen. Läuft der Test von Anfang an durch, müssten wir ja eigentlich gar kein Code mehr schreiben.
 * **Um den Test bestehen zu lassen, nur soviel Code wie nötig schreiben.** Dies unterstreicht einerseits nochmal das Prinzip YAGNI - you ain't gonna need it. Wenn ganz einfacher Code bereits reicht um den Test durchlaufen zu lassen, wieso soll man es sich dann unnötig schwieriger machen. Andererseits zwingt einen dies auch, für jede zusätzliche Komplexität wieder einen neuen Test zu schreiben.
 * **Nach der Implementierung und wenn der Test durchläuft, sollte refactored werden.** Die Änderung für das Bestehen des Test ist möglicherweise noch nicht ganz sauber oder es sind aus früheren Änderungen nun plötzlich bessere und schönere Strukturen im Code erkennbar. Da wir jetzt für alle bisherigen Szenarien Tests haben, können wir problemlos den bestehenden Code verbessern.
